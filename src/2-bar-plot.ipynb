### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['blue', 'green', 'red', 'orange']  # Different colors for each bar

# Create the plot
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title('Bar Plot with Colors')  # Add a title
plt.xlabel('Categories')  # Add x-axis label
plt.ylabel('Values')  # Add y-axis label

# Show the plot
plt.show()
```
