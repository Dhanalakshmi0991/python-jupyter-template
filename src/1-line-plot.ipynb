### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.arange(0, 11, 1)  # x values from 0 to 10
y = x ** 2  # y values as the square of x

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.title('Line Plot of y = xÂ²')  # Add a title
plt.xlabel('x values')  # Add x-axis label
plt.ylabel('y values')  # Add y-axis label
plt.grid(True)  # Add a grid for better readability

# Show the plot
plt.show()
```
